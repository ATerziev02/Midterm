```python
import numpy as np
import pandas as pd

gnrc_data = pd.read_csv('data/GNRC.csv')
holx_data = pd.read_csv('data/HOLX.csv')
mar_data = pd.read_csv('data/MAR.csv')
syy_data = pd.read_csv('data/SYY.csv')
data_list = [gnrc_data, holx_data, mar_data, syy_data]
data_list
```

```python
i = 1
for df in data_list:
    difference = df['Close'].compare(df['Adj Close'])
    if len(difference) == 0:
        print(f"Dataframe number {i} in data_list has same Close and Adj Close ")
   #GNRC and HOLX have the same Close and Adj Close
    
    min = df.min()
    max = df.max()
    print(f"For company {i} the min and max of Close are {min['Close']} and {max['Close']}, whereas the min and max for Adj Close are {min['Adj Close']} and {max['Adj Close']}")
    i = i + 1
```
